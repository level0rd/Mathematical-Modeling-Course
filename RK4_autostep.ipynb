```python
import matplotlib.pyplot as plt
import numpy as np

def f(x, y):
    return -2 * x * y ** 2

def rk4(f, y0, x0, x_end, tol=1e-6):
    h = 0.1
    y = y0
    x = x0
    xs = [x]
    ys = [y]
    while x < x_end:
        k1 = h * f(x, y)
        k2 = h * f(x + h / 2, y + k1 / 2)
        k3 = h * f(x + h / 2, y + k2 / 2)
        k4 = h * f(x + h, y + k3)
        dy = (k1 + 2 * k2 + 2 * k3 + k4) / 6
        if abs(dy) < tol:
            y += dy
            x += h
            xs.append(x)
            ys.append(y)
        else:
            h /= 2
    plt.plot(xs, ys)
    plt.xlabel('x')
    plt.ylabel('y')
    plt.title('Runge-Kutta 4 adaptive step y=-2xy^2')
    plt.show()


y0 = 1
x0 = 0
x_end = 5

rk4(f, y0, x0, x_end)
```
